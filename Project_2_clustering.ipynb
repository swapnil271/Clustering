```python
dataset = pd.read_csv('cell2cellholdout.csv')
```

```python
dataset = dataset.iloc[:,0:20]
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
dataset =  dataset.drop(['Churn'], axis=1)
```

```python
dataset["MonthlyRevenue"].fillna(dataset["MonthlyRevenue"].median(), inplace=True)
dataset["MonthlyMinutes"].fillna(dataset["MonthlyMinutes"].median(), inplace=True)
dataset["TotalRecurringCharge"].fillna(dataset["TotalRecurringCharge"].median(), inplace=True)
dataset["DirectorAssistedCalls"].fillna(dataset["DirectorAssistedCalls"].median(), inplace=True)
dataset["OverageMinutes"].fillna(dataset["OverageMinutes"].median(), inplace=True)
dataset["RoamingCalls"].fillna(dataset["RoamingCalls"].median(), inplace=True)
dataset["PercChangeMinutes"].fillna(dataset["PercChangeMinutes"].median(), inplace=True)
dataset["PercChangeRevenues"].fillna(dataset["PercChangeRevenues"].median(), inplace=True)
```

```python
dataset.isnull().sum()
```

```python
dataset
```

```python
x = dataset.iloc[:,:].values
x
```

```python
# K-Mean_Techiques
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0)
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
datset_new = pd.concat([dataset, pd.DataFrame(y_kmeans)], axis=1)
datset_new
```

```python
plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1], s=100, c='red', label='cluster1')
plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1], s=100, c='pink', label='cluster2')
plt.scatter(x[y_kmeans==2,0], x[y_kmeans==2,1], s=100, c='yellow', label='cluster3')
plt.scatter(x[y_kmeans==3,0], x[y_kmeans==3,1], s=100, c='black', label='cluster4')
plt.scatter(x[y_kmeans==4,0], x[y_kmeans==4,1], s=100, c='blue', label='cluster4')
```

```python
#Cluster_Hierarchy
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(x,method='ward'))

```

```python
plt.title('Dendrogram')
```

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5, affinity='euclidean', 
                             linkage='ward')
```

```python
y_hc = hc.fit_predict(x)
```

```python
y_hc
```

```python
plt.scatter(x[y_hc==0,0], x[y_hc==0,1], s=100, c='red', label='cluster1')
plt.scatter(x[y_hc==1,0], x[y_hc==1,1], s=100, c='pink', label='cluster1')
plt.scatter(x[y_hc==2,0], x[y_hc==2,1], s=100, c='yellow', label='cluster1')
plt.scatter(x[y_hc==3,0], x[y_hc==3,1], s=100, c='black', label='cluster1')
plt.scatter(x[y_hc==4,0], x[y_hc==4,1], s=100, c='green', label='cluster1')
```
