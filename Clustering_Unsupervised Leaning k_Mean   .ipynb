```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
x = dataset.iloc[:, [3,4]].values
x
```

```python
# K-Mean_Techiques
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++')
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11), wcss)
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++')
y_kmeans = kmeans.fit_predict(x)
```

```python
y_kmeans
```

```python
dataset
```

```python
y_kmeans_d = pd.DataFrame(y_kmeans).rename(columns={0:"Outcomes"})
y_kmeans_d
```

```python
dataset_new = pd.concat([dataset, pd.DataFrame(y_kmeans_d)], axis=1)
dataset_new
```

```python
plt.scatter(x[y_kmeans==0,0], x[y_kmeans==0,1], s=100, c='red', label='cluster1')
plt.scatter(x[y_kmeans==1,0], x[y_kmeans==1,1], s=100, c='pink', label='cluster1')
plt.scatter(x[y_kmeans==2,0], x[y_kmeans==2,1], s=100, c='yellow', label='cluster1')
plt.scatter(x[y_kmeans==3,0], x[y_kmeans==3,1], s=100, c='black', label='cluster1')
plt.scatter(x[y_kmeans==4,0], x[y_kmeans==4,1], s=100, c='green', label='cluster1')
```
